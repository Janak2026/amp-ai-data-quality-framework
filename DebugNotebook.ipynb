```python
import os

def tree(path, indent=""):
    for item in sorted(os.listdir(path)):
        full = os.path.join(path, item)
        print(indent + "├── " + item)
        if os.path.isdir(full):
            tree(full, indent + "│   ")

tree("/Workspace/Repos/jana21.amazon@gmail.com/amp-ai-data-quality-framework")
```
